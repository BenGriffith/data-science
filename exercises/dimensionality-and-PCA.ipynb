### Create a new score variable from the weighted averages of all score variables in the datasets. Notice that the number of students in the 4th grade isn't the same as the number of students in the 8th grade. So, you should appropriately weigh the scores! 

````python
edu_df['TOTAL_SCORE'] = (edu_df['GRADES_4_G'] * ((edu_df['AVG_MATH_4_SCORE'] + edu_df['AVG_READING_4_SCORE']) * 0.5)
                          +
                           edu_df['GRADES_8_G'] * ((edu_df['AVG_MATH_8_SCORE'] + edu_df['AVG_READING_8_SCORE']) * 0.5)
                          ) / (edu_df['GRADES_4_G'] + edu_df['GRADES_8_G'])
````

### What are the correlations between this newly created score variable and the expenditure types? Which 1 of the expenditure types is more correlated than the others? 

````python
edu_df[['TOTAL_SCORE', 'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE', 'OTHER_EXPENDITURE', 'CAPITAL_OUTLAY_EXPENDITURE']].corr()
````

Based on the correlation matrix above, ```edu_df['INSTRUCTION_EXPENDITURE']``` is mroe correlated than the others.

### Now, apply PCA to the 4 expenditure types. How much of the total variance is explained by the 1st component? 

````python
X = StandardScaler().fit_transform(edu_df[['INSTRUCTION_EXPENDITURE', 
                                           'SUPPORT_SERVICES_EXPENDITURE', 
                                           'OTHER_EXPENDITURE', 
                                           'CAPITAL_OUTLAY_EXPENDITURE']])

sklearn_pca = PCA(n_components=4)
Y_sklearn = sklearn_pca.fit_transform(X)

print('The percentage of total variance in the expenditure types explained by each',
      'component from Sklearn PCA.\n', sklearn_pca.explained_variance_ratio_)
````

````python
X1 = StandardScaler().fit_transform(edu_df[['INSTRUCTION_EXPENDITURE', 
                                            'SUPPORT_SERVICES_EXPENDITURE', 
                                            'OTHER_EXPENDITURE', 
                                            'CAPITAL_OUTLAY_EXPENDITURE']])

sklearn_pca1 = PCA(n_components=1)
edu_df['PCA'] = sklearn_pca1.fit_transform(X1)

print('The percentage of total variance in the expenditure types explained by each',
      'component from Sklearn PCA.\n', sklearn_pca1.explained_variance_ratio_)
````
